# Python Básico

## Operações com Númericos

Operadores:


*   Soma: +
*   Subtração: -
*   Divisão: /
*   Multiplicação: *
*   Potência: **
*   Módulo: %
*   Divisão inteira: //



```python
# Exemplo de uso dos operadores

num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))

# Exibir resultados
print("\nResultados:")
print(f"{num1} + {num2} =", num1 + num2)
print(f"{num1} - {num2} =", num1 - num2)
print(f"{num1} * {num2} =", num1 * num2)
print(f"{num1} / {num2} =", num1 / num2)
print(f"{num1} ** {num2} =", num1 ** num2)
print(f"{num1} // {num2} =", num1 // num2)
print(f"{num1} % {num2} =", num1 % num2)
```

## Operações com Strings

Operadores:

*  Concatenação
*  Repetição
*  Indexação
*  Fatiamento (substring)
*  Comprimento (Comprimento)
*  Verificação de substring
*  Métodos de String  

```python
# Strings de exemplo
string1 = "Olá, "
string2 = "Mundo!"

# 1. Concatenação (+)
print("   Exemplo: string1 + string2 =", string1 + string2)
print()

# 2. Repetição (*)
print("   Exemplo: string1 * 3 =", string1 * 3)
print()

# 3. Indexação
print("   Exemplo: string2[1] =", string2[1])  # Acessa o segundo caractere (índice 1)
print()

# 4. Fatiamento (slicing)
print("   Exemplo: string2[1:4] =", string2[1:4])  # Retorna os caracteres do índice 1 ao 3
print()

# 5. Comprimento (len)
print("   Exemplo: len(string2) =", len(string2))  # Retorna o número de caracteres na string
print()

# 6. Verificação de substring (in)
print("   Exemplo: 'Mun' in string2 =", 'Mun' in string2)  # Verifica se 'Mun' está em string2
print()

# 7. Métodos de strings
print("7. Métodos de strings")
print("   a. string2.upper() =", string2.upper())  # Converte para maiúsculas
print("   b. string2.lower() =", string2.lower())  # Converte para minúsculas
print("   c. string2.replace('Mundo', 'Python') =", string2.replace('Mundo', 'Python'))  # Substitui uma substring
print("   d. string2.split('u') =", string2.split('u'))  # Divide a string em uma lista com base no caractere 'u'
print("   e. string2.strip('!') =", string2.strip('!'))  # Remove caracteres especificados das extremidades
print()

# 8. Formatação de strings (f-strings)
nome = "João"
idade = 25
print(f"   Exemplo: f'Meu nome é {nome} e tenho {idade} anos.' =", f'Meu nome é {nome} e tenho {idade} anos.')
print()
```

## Operações com Listas

```python
# Criando uma lista de exemplo
lista = [1, 2, 3, 4, 5]

# 1. Acessando elementos da lista
print("   Lista original:", lista)
print("   Primeiro elemento (índice 0):", lista[0])
print("   Último elemento (índice -1):", lista[-1])
print()

# 2. Modificando elementos da lista
lista[2] = 10  # Altera o terceiro elemento (índice 2)
print("   Lista após modificar o terceiro elemento:", lista)
print()

# 3. Adicionando elementos à lista
lista.append(6)  # Adiciona um elemento ao final da lista
print("   Lista após append(6):", lista)
lista.insert(1, 7)  # Insere o número 7 no índice 1
print("   Lista após insert(1, 7):", lista)
print()

# 4. Removendo elementos da lista
lista.remove(10)  # Remove o primeiro elemento com valor 10
print("   Lista após remove(10):", lista)
elemento_removido = lista.pop(1)  # Remove e retorna o elemento no índice 1
print("   Lista após pop(1):", lista)
print("   Elemento removido com pop(1):", elemento_removido)
print()

# 5. Fatiamento (slicing)
print("   Lista original:", lista)
print("   Elementos do índice 1 ao 3:", lista[1:4])
print("   Todos os elementos a partir do índice 2:", lista[2:])
print("   Todos os elementos até o índice 3:", lista[:4])
print()

# 6. Comprimento da lista
print("   Número de elementos na lista:", len(lista))
print()

# 7. Ordenação da lista
lista.sort()  # Ordena a lista em ordem crescente
print("   Lista ordenada em ordem crescente:", lista)
lista.sort(reverse=True)  # Ordena a lista em ordem decrescente
print("   Lista ordenada em ordem decrescente:", lista)
print()

# 8. Inversão da lista
lista.reverse()  # Inverte a ordem dos elementos da lista
print("   Lista invertida:", lista)
print()

# 9. Verificação de existência de um elemento
print("   O número 5 está na lista?", 5 in lista)
print("   O número 10 está na lista?", 10 in lista)
print()

# 10. Concatenando listas
outra_lista = [8, 9, 10]
lista_concatenada = lista + outra_lista
print("   Lista concatenada:", lista_concatenada)
print()

# 11. Limpando a lista
lista.clear()
print("   Lista após clear():", lista)
print()
```

## Loops FOR

```python
# 1. Loop for com range
print("1. Loop for com range:")
# range(5) gera números de 0 a 4
for i in range(5):
    print("   Número do range:", i)

# range com início, fim e passo
print("\n   range(2, 10, 2):")
for i in range(2, 10, 2):  # Começa em 2, vai até 10 (exclusivo), passo 2
    print("   Número do range:", i)

# 2. Loop for em listas
print("\n2. Loop for em listas:")
lista = ["Maçã", "Banana", "Laranja", "Uva"]

# Iterando sobre os elementos da lista
print("   Iterando sobre os elementos:")
for fruta in lista:
    print("   Fruta:", fruta)

# Iterando sobre os índices e elementos da lista
print("\n   Iterando sobre índices e elementos:")
for indice, fruta in enumerate(lista):
    print(f"   Índice {indice}: {fruta}")

# 3. Loop for em strings
print("\n3. Loop for em strings:")
texto = "Python"

# Iterando sobre os caracteres da string
print("   Iterando sobre os caracteres:")
for caractere in texto:
    print("   Caractere:", caractere)

# Iterando sobre os índices e caracteres da string
print("\n   Iterando sobre índices e caracteres:")
for indice, caractere in enumerate(texto):
    print(f"   Índice {indice}: {caractere}")

# 4. Loop for com listas aninhadas
print("\n4. Loop for com listas aninhadas:")
matriz = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Iterando sobre uma matriz (lista de listas)
print("   Iterando sobre uma matriz:")
for linha in matriz:
    for elemento in linha:
        print("   Elemento:", elemento)

# 5. Loop for com dicionários
print("\n5. Loop for com dicionários:")
dicionario = {"nome": "João", "idade": 25, "cidade": "São Paulo"}

# Iterando sobre as chaves do dicionário
print("   Iterando sobre as chaves:")
for chave in dicionario:
    print("   Chave:", chave)

# Iterando sobre os valores do dicionário
print("\n   Iterando sobre os valores:")
for valor in dicionario.values():
    print("   Valor:", valor)

# Iterando sobre chaves e valores do dicionário
print("\n   Iterando sobre chaves e valores:")
for chave, valor in dicionario.items():
    print(f"   {chave}: {valor}")

# 6. Loop for com else
print("\n6. Loop for com else:")
# O bloco else é executado após o loop for ser concluído
for i in range(3):
    print("   Número:", i)
else:
    print("   Loop for concluído!")

# 7. Loop for com break e continue
print("\n7. Loop for com break e continue:")
# Exemplo de break: interrompe o loop quando encontra o número 3
print("   Exemplo de break:")
for i in range(5):
    if i == 3:
        print("      Encontrei o número 3! Parando o loop.")
        break
    print("      Número:", i)

# Exemplo de continue: pula a iteração quando o número é par
print("\n   Exemplo de continue:")
for i in range(5):
    if i % 2 == 0:
        continue  # Pula para a próxima iteração
    print("      Número ímpar:", i)
```

## Loops While

```python
# Exemplos de uso do loop while

# 1. Loop while básico
print("1. Loop while básico:")
contador = 0
while contador < 5:
    print("   Contador:", contador)
    contador += 1  # Incrementa o contador

# 2. Loop while com condição complexa
print("\n2. Loop while com condição complexa:")
numero = 10
while numero > 0 and numero % 2 == 0:  # Continua enquanto o número for positivo e par
    print("   Número par positivo:", numero)
    numero -= 2  # Decrementa o número em 2

# 3. Loop while com break
print("\n3. Loop while com break:")
contador = 0
while True:  # Loop infinito
    print("   Contador:", contador)
    contador += 1
    if contador == 5:  # Condição para sair do loop
        print("   Contador atingiu 5. Saindo do loop.")
        break

# 4. Loop while com continue
print("\n4. Loop while com continue:")
contador = 0
while contador < 10:
    contador += 1
    if contador % 2 == 0:  # Pula números pares
        continue
    print("   Número ímpar:", contador)

# 5. Loop while com else
print("\n5. Loop while com else:")
contador = 0
while contador < 3:
    print("   Contador:", contador)
    contador += 1
else:
    print("   Loop while concluído!")  # Executado quando a condição do while se torna falsa

# 6. Loop while para validação de entrada
print("\n6. Loop while para validação de entrada:")
while True:
    entrada = input("   Digite um número maior que 10: ")
    if entrada.isdigit():  # Verifica se a entrada é um número
        numero = int(entrada)
        if numero > 10:
            print("   Número válido:", numero)
            break
        else:
            print("   Número inválido. Tente novamente.")
    else:
        print("   Entrada inválida. Digite um número.")

# 7. Loop while com listas
print("\n7. Loop while com listas:")
lista = [1, 2, 3, 4, 5]
while lista:  # Continua enquanto a lista não estiver vazia
    elemento = lista.pop()  # Remove e retorna o último elemento
    print("   Elemento removido:", elemento)
print("   Lista após o loop:", lista)

# 8. Loop while aninhado
print("\n8. Loop while aninhado:")
linha = 1
while linha <= 3:
    coluna = 1
    while coluna <= 3:
        print(f"   Linha {linha}, Coluna {coluna}")
        coluna += 1
    linha += 1

# 9. Loop while com decremento
print("\n9. Loop while com decremento:")
contador = 5
while contador > 0:
    print("   Contador regressivo:", contador)
    contador -= 1  # Decrementa o contador

# 10. Loop while com manipulação de strings
print("\n10. Loop while com manipulação de strings:")
texto = "Python"
indice = 0
while indice < len(texto):
    print("   Caractere:", texto[indice])
    indice += 1
```

## Funções e Classes

```python
# 1. Funções
# Definição de uma função simples
def saudacao(nome):
    return f"Olá, {nome}!"

# Chamando a função
mensagem = saudacao("João")
print("   Exemplo de função:", mensagem)

# Função com múltiplos parâmetros e valor padrão
def calcular_area_retangulo(largura, altura=10):
    return largura * altura

# Chamando a função com e sem valor padrão
print("   Área do retângulo (5x10):", calcular_area_retangulo(5))
print("   Área do retângulo (5x20):", calcular_area_retangulo(5, 20))

# Função com número variável de argumentos
def somar(*numeros):
    # print(type(numeros))
    return sum(numeros)

# Chamando a função com múltiplos argumentos
print("   Soma de 1, 2, 3, 4, 5:", somar(1, 2, 3, 4, 5))

```

```python
# 1. Classes, atributos e métodos

# Definição de uma classe simples
class Carro:
    # Atributo de classe (compartilhado por todas as instâncias)
    rodas = 4

    # Método construtor (inicializa atributos de instância)
    def __init__(self, marca, modelo, ano):
        self.marca = marca  # Atributo de instância
        self.modelo = modelo  # Atributo de instância
        self.ano = ano  # Atributo de instância

    # Método de instância
    def descricao(self):
        return f"{self.marca} {self.modelo} ({self.ano})"

    # Método de classe
    @classmethod
    def quantidade_rodas(cls):
        return f"Todos os carros têm {cls.rodas} rodas."

    # Método estático (não depende de instância ou classe)
    @staticmethod
    def buzinar():
        return "Biiiiiiii!"

# Criando uma instância da classe Carro
meu_carro = Carro("Toyota", "Corolla", 2022)

# Acessando atributos de instância
print("   Marca do carro:", meu_carro.marca)
print("   Modelo do carro:", meu_carro.modelo)
print("   Ano do carro:", meu_carro.ano)

# Chamando métodos de instância
print("   Descrição do carro:", meu_carro.descricao())

# Acessando atributo de classe
print("   Quantidade de rodas:", Carro.rodas)

# Chamando método de classe
print("   Informação sobre rodas:", Carro.quantidade_rodas())

# Chamando método estático
print("   Buzinando:", Carro.buzinar())

print('\n'*2)

# 2. Encapsulamento e propriedades
class ContaBancaria:
    def __init__(self, titular, saldo=0):
        self.titular = titular
        self.__saldo = saldo  # Atributo privado (encapsulado)

    # Propriedade para acessar o saldo
    @property
    def saldo(self):
        return self.__saldo

    # Método para depositar dinheiro
    def depositar(self, valor):
        if valor > 0:
            self.__saldo += valor
            print(f"   Depósito de R${valor} realizado.")
        else:
            print("   Valor de depósito inválido.")

    # Método para sacar dinheiro
    def sacar(self, valor):
        if 0 < valor <= self.__saldo:
            self.__saldo -= valor
            print(f"   Saque de R${valor} realizado.")
        else:
            print("   Saldo insuficiente ou valor inválido.")

# Criando uma instância da classe ContaBancaria
minha_conta = ContaBancaria("Maria", 1000)

# Acessando a propriedade saldo
print("   Saldo inicial:", minha_conta.saldo)

# Realizando operações na conta
minha_conta.depositar(500)
minha_conta.sacar(200)
print("   Saldo final:", minha_conta.saldo)
```

## Bibliotecas

**O que são Bibliotecas?**

Bibliotecas (ou módulos) são coleções de código organizadas em arquivos ou pacotes que podem ser reutilizados em diferentes projetos.

Elas podem conter **funções, classes, variáveis** e outros recursos.

Bibliotecas são usadas para **evitar a repetição de código e para aproveitar funcionalidades já implementadas por outras pessoas**.

# Biblioteca Pandas

```python
# Criando um DataFrame a partir de um dicionário
data = {
    'Nome': ['Alice', 'Bob', 'Carlos', 'Daniela'],
    'Idade': [25, 30, 35, 40],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Porto Alegre'],
    'Salário': [5000, 6000, 7000, 8000]
}
```
